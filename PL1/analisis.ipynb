#### $R_C=1k\Omega$

```python
# Paquetes necesarios

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
V_BB = np.array([1.07, 2.01, 2.95, 4.02, 5.01, 5.89, 6.98, 7.43, 8.00, 8.49, 8.96, 9.86, 11.98, 14.88, 17.40, 19.94])
V_CE = np.array([14.60, 13.54, 12.50, 11.24, 10.11, 9.13, 7.91, 7.43, 6.80, 6.28, 5.83, 4.90, 2.97, 0.8, 0.37, 0.257])
V_R_C = np.array([0.48, 1.53, 2.56, 4.96, 5.95, 7.15, 7.61, 8.26, 8.77, 9.24, 10.15, 12.00, 14.30, 14.70, 14.70, 14.82])

I_C = V_R_C

regresion = linregress(V_CE, I_C)
fit = regresion.slope * V_CE + regresion.intercept

plt.plot(V_CE, I_C)
plt.plot(V_CE, fit)
plt.xlabel("$V_{CE}$ (V)")
plt.ylabel("$I_C$ (mA)")
plt.title(r"Recta de carga, $R_C=1k\Omega$")
plt.text(12, 14, f"R^2={regresion.rvalue**2:.2f}")
plt.show()
```
